```python
import h5py
import numpy as np
import os
```

```python
directory = '/storage/home/hcoda1/0/jw254/data/RS-RP/May_2018-fesc/RP/fesc-data'
mass_all_halo = []
N_tot_all_halo = []
pos_all_halo = []

def sorter(x):
    return(x[15:19])

for i in sorted(os.listdir(directory), key = sorter):
    print(i)
    if i.endswith('.h5'):
        with h5py.File(os.path.join(directory, i)) as data:
            for halo in data:
                mass = []
                N_tot = []
                pos = []
                if data[halo]['fesc'][()] >= 1e-5:
                    for star in data[halo]['Pop2_Stars'].keys():
                        mass.append(data[halo]['Pop2_Stars'][star]['mass'][()])
                        N_tot.append(data[halo]['Pop2_Stars'][star]['Ntot'][()])
                        pos.append(data[halo]['Pop2_Stars'][star]['pos'][()])
                    mass_all_halo.append(mass)
                    N_tot_all_halo.append(N_tot)
                    pos_all_halo.append(pos)
            
        
            
```

```python
np.save('mass.npy', mass_all_halo)
np.save('Ntot.npy', N_tot_all_halo)
np.save('pos.npy', pos_all_halo)
```
